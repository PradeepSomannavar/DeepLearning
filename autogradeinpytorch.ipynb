```python
import torch
```

```python
material_cost = torch.tensor( 10, requires_grad = True,dtype = torch.float16)
```

```python
material_cost
```

```python
labour_cost = torch.tensor( 5, requires_grad = True,dtype = torch.float16)
```

```python
total_cost = 3 * material_cost**2 + 5 * labour_cost + 100
```

```python
total_cost
```

```python
total_cost.backward()
```

```python
material_cost.grad
```

```python
labour_cost.grad
```

```python
x = torch.tensor( 4, requires_grad = True,dtype = torch.float16)
```

```python
y = x ** 2 + 5
```

```python
y.backward()
```

```python
x.grad
```
